- uniform dağılım
    - belirli bir aralıkta [a,b] tüm değerler eşit olasılıkla ortaya çıkar
    - PDF düz bir şekildedir
    - zar atıldığında her sonucun eşit olasılıkla gelmesi
- norma dağılım (gauss)
    - veriler ortalamaya yakın değerlerde yoğunlaşır ve uç değerlere doğru simetrik olarak azalır
    - PDF çan eğrisi şeklindedir.
    - sınav sonuçları, insan boyları

#### olasılık dağılım fonksiyonları





```python
import matplotlib.pyplot as plt 

# Uniform dağılımdan örnekler üret  
uniform_veri = np.random.uniform(0, 10, 1000)  
plt.hist(uniform_veri, bins=30, density=True)  
plt.title("Uniform Dağılım")  
plt.show()  
```

```python
# Normal dağılımdan örnekler üret  
normal_veri = np.random.normal(5, 2, 1000)  # Ortalama=5, Standart Sapma=2  
plt.hist(normal_veri, bins=30, density=True)  
plt.title("Normal Dağılım")  
plt.show()  
```

#### filtering

```python
arr = np.array([[0, 2, np.nan],
                [1, np.nan, -6],
                [np.nan, -2, 1]])
print(repr(np.isnan(arr)))
```

```python
arr = np.array([0, 3, 5, 3, 1])
print(np.where(arr==3))
```

```python
arr = np.array([[-2, -1, -3],
                [4, -5, -6],
                [3, -9, 1]])
print(arr>0)
```

```python
# axis = 0 fonksiyon column üzerinde çalışır
# axis = 1 fonksiyon satirlar üzerinde çalışır
print(np.any(arr>3,axis=0))
print(np.any(arr>3,axis=1))
```

```python
print(repr(np.all(arr > 0, axis=1)))
print(repr(np.all(arr > 3, axis=0)))
```

```python
arr = np.array([[-2, -1, -3],
                [4, 5, -6],
                [3, 9, 1]])
has_positive = np.any(arr > 0, axis=1)
print(has_positive)
print(repr(arr[np.where(has_positive)]))
```

### statistics

#### analysis
- min, max vs gib fonksiyonlar




```python
arr = np.array([[0, 72, 3],
                [1, 3, -60],
                [-3, -2, 4]])
print(arr.min())
print(arr.max())

print(repr(arr.min(axis=0)))
print(repr(arr.max(axis=-1)))
```

#### metrics

```python
arr = np.array([[0, 72, 3],
                [1, 3, -60],
                [-3, -2, 4]])
print(np.mean(arr))
print(np.var(arr))
print(np.median(arr))
print(repr(np.median(arr, axis=-1)))
```

### aggregation

```python
arr = np.array([[0, 72, 3],
                [1, 3, -60],
                [-3, -2, 4]])
print(np.sum(arr))
print(repr(np.sum(arr, axis=0)))
print(repr(np.sum(arr, axis=1)))
```

```python
arr = np.array([[1, 2 ],
                [4, 5 ],
                [6,7]]) 
print(repr(np.cumsum(arr)))
print(repr(np.cumsum(arr, axis=0)))
print(repr(np.cumsum(arr, axis=1)))
```
